<h3> 1. Introduction</h3>

> <h4> 1.1 Définition </h4>
>
> L'**apprentissage automatique**, dit aussi apprentissage machine ou **machine learning** en anglais, est un type d'intelligence artificielle. Cette intelligence artificielle apprend à partir de données qui lui sont fournies et cet apprentissage va lui permettre de prendre des décisions. 
> On parle plus communément de modèles de Machine Learning : on fait alors référence à l'entraînement d'un algorithme sur des données. On aura souvent pour objectif d'utiliser ce modèle afin de réaliser des prédictions sur de nouvelles données, inconnues par le modèle. 
>

> <h4> 1.2 Apprentissage supervisé ou non supervisé </h4>
>
> Le Machine Learning regroupe deux grands types de problématiques d'apprentissage :
>
> - L'<span style="color:#4529de"> <b>apprentissage supervisé </b> </span> ou <span style="color:#4529de"> <b>supervised learning</b> </span> : C'est lorsqu'il y a une variable cible. A partir de données labellisées (dites aussi étiquettées), on cherche à **prédire** cette variable. On a à disposition des variables explicatives et on veut prédire l'information contenue dans la variable cible. Les données disponibles sont labellisées, cela signifie que l'on connaît la valeur de la variable cible pour ces données. Cela est nécessaire pour permettre au modèle de s'entraîner : apprendre sur les données.
>
> L'apprentissage <span style="color:#4529de"> <b>supervisé</b> </span> peut être subdivisé en deux principales techniques :
>
> * La <span style="color:#4529de"> <b>régression</b> </span> : On cherche à prédire une variable de type quantitatif.
> <br/>
> <center><video controls src="https://assets-datascientest.s3.eu-west-1.amazonaws.com/notebooks/module_da_121/notebook_introduction/reg_gif.mov" width = '500' /></center> 
> <br/>
>
> * La <span style="color:#4529de"> <b>classification </b> </span> : On cherche à prédire une variable de type qualitatif.
> <br/>
> <center><video controls src="https://assets-datascientest.s3.eu-west-1.amazonaws.com/notebooks/module_da_121/notebook_introduction/classif_gif.mov" width = '500' /></center> 
> <br/>
> - L'<span style="color:#a329de"> <b>apprentissage non supervisé </b> </span> ou <span style="color:#a329de"> <b>unsupervised learning</b> </span> : C'est lorsqu'il n'y a pas de variable cible : on cherche à découvrir les **structures sous-jacentes** au sein d'une base de données.
> 
>
> L'apprentissage <span style="color:#a329de"> <b>non supervisé</b> </span> peut être subdivisé en deux principales techniques:
>
> * La <span style="color:#a329de"> <b>réduction de dimension</b> </span>: On transforme nos variables pour réduire le nombre de colonnes dans nos données. 
>
>
> * Le <span style="color:#a329de"> <b>clustering </b> </span> : On cherche à regrouper nos observations au sein de groupes.
> <br/> <br/>

<h3> 2. Pourquoi faire de la modélisation ? </h3>

> <h4> 2.1 Un outil d'aide à la décision </h4>
>
> L'utilisation du Machine Learning peut se révéler très intéressante au sein des entreprises dans le cadre du développement d'outils d'aide à la prise de décision. Le développement de modèles, tels que des modèles de marketing prédictif, de reconnaissance faciale ou encore de détection de fraude, permet de faciliter le traitement de certaines problématiques mais aussi d'appuyer les décisions des entreprises d'un point de vue financier ou stratégique. 
>
> <h4> 2.2 La pipeline de traitement de la donnée </h4>
>
>

>  Il est important de garder en tête que les analyses et applications de modèles de Machine Learning ne sont que les étapes finales par lesquelles passent les données. En amont, ces mêmes données auront subies plusieurs transformations.
> 
> Suite à la définition des besoins, les étapes de collecte de la donnée doivent être mises en place. On parle de **Data Collection**.
> A titre d'exemple, l'utilisation de techniques de webscraping peuvent nous permettre de récupérer de la donnée sur des sites internet. A l'issue de cette étape nous aurons une ou plusieurs **sources de données**.
>
> Ces mêmes données devront alors être récupérées et ajoutées à un **site de destination**, on parle alors de **Data Ingestion**. Le site de destination pourra être une base de données ou un entrepôt de données. Durant cette étape il conviendra de réfléchir à la structure que l'on souhaite faire adopter à nos données. Cette structure aura une importance capitale dans la suite du processus du traitement de la donnée. Cette étape d'ingestion englobe les **étapes ETL (Extract, Transform and Load)** : les données sont extraites, transformées et finalement chargées dans le site de destination final. Ces étapes sont **primordiales** car elles garantissent un bon niveau de qualité de données.
> Une fois ces étapes finalisées, les métiers peuvent alors récupérer les données nécessaires à leurs analyses et projets pour les traiter.
>

<h3> 3. Evaluation des modèles et performance </h3>

> <h4> 3.1 La notion de performance </h4>
>
> Nous abordons dans ce premier notebook, la notion de performance de modèle puisqu'elle est au coeur de la problématique d'apprentissage supervisé. La mesure de performance permet d'évaluer le caractère opérationnel des modèles. En effet, plus **un modèle est performant** plus il est **fiable** et donc utilisable. 
>
>A titre d'exemple, on peut imaginer un modèle qui **classe des emails**. Le but est d'**identifier quels sont les spams**. Si notre modèle est **peu performant** alors des emails non frauduleux pourraient être mal classés et inversement : l'utilisation de notre modèle est ici contre-productive puisqu'il ne nous aura pas permis d'identifier parmi tous les emails ceux qui sont frauduleux.
>
> Il est à noter que la performance est déterminée en grande partie par la qualité des données. Il faut donc veiller à ce que nos données soient bien traitées avant de passer à l'étape de modélisation.
>
> Plus concrètement, pour connaître la performance et donc décider qu'un modèle est opérationnel, **on mesure à l'aide de métriques la qualité de la prédiction en comparant les valeurs prédites aux valeurs réelles sur les données à notre disposition pour réaliser le modèle**.
>
> Le réalisation du modèle est une étape délicate car il faut veiller à ce qu'il soit généralisable. Bien que la mesure des performances d'un modèle se fasse sur les données que nous avons à disposition, l'objectif est qu'il soit **performant sur de nouvelles données** dont on ne connaît pas l'issue (la variable cible). 
> Dans le cas où le modèle performe uniquement sur les données qu'il connait mais pas sur de nouvelles données, on parle de **surapprentissage**. Le modèle est alors **trop adapté** aux données sur lesquelles il s'est entraîné. 
>
> <h4> 3.2 La technique simple de hold-out : Train - Test - Split </h4>
>
>Pour être certain de ne pas se trouver dans ce genre de cas, la performance de prédiction des modèles doit être évaluée sur une partie des données qui n'a pas été observée par le modèle. Pour cela, on divise les données en deux parties, une partie dédiée à l'entraînement et une autre dédiée à l'évaluation. On obtient alors les **performances prédictives réelles** d'un modèle. Ci-après la typologie utilisée et les définitions associées à chaque terme:
>
>* <span style="color:#4529de"> <b> Jeu d'entraînement (Train set) : </b>  </span><br> C'est la partie du jeu de données sur laquelle le modèle va apprendre et s'entraîner. Les paramètres du modèle vont donc être estimés durant cette étape.
>
>
>* <span style="color:#4529de"> <b> Jeu de test (Test set) : </b>  </span><br>  C'est la partie du jeu de données qui n'a jamais été étudiée par le modèle, c'est sur cette partie des données que les performances seront évaluées. 
>
><div class="alert alert-warning">
    <i class="fa fa-info-circle"></i> &emsp; Le jeu d'entraînement du modèle représente souvent 70 à 80% du volume total des données pour un jeu de test représentant 20 à 30%. Il n'y a néanmoins pas de règle générale, ces proportions peuvent varier selon la nature et le volume des données. </div>

> <h4>3.4 Déploiement de modèle </h4>
>
> Une fois le modèle entraîné et évalué, si un modèle est jugé opérationnel, vient alors l'étape de mise en production. Une fois mis en production les modèles peuvent être utilisés à l'échelle de l'entreprise.
> Le modèle est alors utilisé pour réaliser des prédictions sur des nouvelles données. Il pourra alors avoir un rôle crucial dans les prises de décision des équipes.

<h3> 4. Conclusion </h3>

> <h4>4.1 A retenir </h4>
>
> 
> * L'**apprentissage supervisé** se distingue de l'apprentissage **non supervisé** : on cherche dans le premier cas à prédire une **variable cible prédéfinie**, dans le second à trouver au sein des données **non labellisées des structures existantes**. 
>
>
> * On parle de **régression** lorsque l'on cherche à prédire une **variable quantitative** et de **classification** lorsque l'on cherche à prédire une **variable qualitative**.
>
>
> * La séparation du jeu de données, en un jeu d'entraînement et un jeu de test, permet d'évaluer les **performances** et la **capacité de généralisation** d'un modèle.
>
>
> * L'évaluation des performances d'un modèle sur des données qui n'ont jamais été entraînées est **essentielle**, cela permet notamment de décéler un problème de surapprentissage. Pour cela on compare les performances du modèle sur le jeu d'entraînement et de test.
