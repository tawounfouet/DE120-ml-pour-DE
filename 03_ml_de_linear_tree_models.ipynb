<h3> 1. Introduction </h3> 

> Dans ce notebook nous allons étudier les **modèles de type arbre** et les **modèles linéaires**. Nous choisissons d'étudier ces modèles car ils sont simples et interprétables. Il est néanmoins à noter qu'il existe une multitude d'autres modèles. On peut alors naturellement se demander quel serait l'avantage d'en utiliser un plutôt qu'un autre. <br><br>
> Dans le précédent notebook, il était précisé que l'entraînement des modèles de type arbre nécessitait peu ou pas de preprocessing. Ce genre d'argument peut bien entendu nous pousser à utiliser un algortihme plutôt qu'un autre. Mais il existe d'autres raisons qui déterminent le choix des algorithmes.<br><br>
> Lorsque l'on utilise un modèle linéaire, on suppose qu'il existe une **relation linéaire** entre nos variables explicatives et notre variable cible. Très intuitifs et fiables, ces modèles sont plébiscités dans de nombreux secteurs. 
> En revanche, lorsque les relations entre les variables sont **plus complexes**, l'utilisation de ce genre d'algorithmes ne permet pas d'aboutir à des modèles à forte performance. Il faut alors priviliégier des modèles **non-linéaires** parmi lesquels on peut compter les modèles de type arbre.<br><br>

> Les modèles de type arbre ont aussi leurs limites, en particulier, les arbres de décision simples sont sujet au **surapprentissage**. Cela signifie que le modèle sera **"suradapté" aux données d'entraînement** et qu'il perfomera difficilement sur de nouvelles données. Des techniques existent néanmoins pour pallier ce problème. 

<h3> 2. Les arbres de décision </h3> 

><h4> 2.1 Présentation </h4> 
>
>
> Un arbre de décision est un modèle de Machine Learning utilisé à la fois en **régression** et en **classification**. Ce modèle cherche à séparer les échantillons en groupes d’échantillons les plus “homogènes” possible par rapport à la variable cible. Plus ils sont homogènes, plus le modèle est performant. <br><br>
> Pour les séparer, il va effectuer une série de tests sur les différentes variables du jeu de données. Ces étapes sont représentées par des nœuds dans l’arbre.<br><br>
> L’arbre de décision possède plusieurs éléments : des nœuds (dont le nœud racine), des branches et des feuilles qui sont les nœuds terminaux aux extrémités de l’arbre. 
>
> > * Chaque <b>nœud</b> correspond à un test sur une variable.
> >
> >
> > * Chacune des <b>branches</b> représente un résultat de ce test.
> >
> >
> > * Chaque <b>feuille</b> de l'arbre (nœud final) indique la valeur prédite de la variable cible.
> >
> > L'arbre se lit de haut en bas, niveau par niveau.
> >
> > <img src="https://assets-datascientest.s3.eu-west-1.amazonaws.com/notebooks/module_da_121/notebook3/decision_tree.png" style="height:400px"> 
>
><h4> 2.2 L'arbre de régression </h4> 
>
> Pour mieux comprendre leur construction, nous allons prendre en exemple un arbre de décision pour un problème de régression. 
> On cherche dans notre cas à prédire le montant des indemnités (Y, la variable cible) offert par une compagnie d'assurance à partir du nombre de réclamations réalisées (X, la variable explicative). 
Voici l'arbre de décision que l'on obtient après entraînement du modèle : 
> <img src="https://assets-datascientest.s3.eu-west-1.amazonaws.com/notebooks/module_da_121/notebook3/arbre_de_regression.png" style="height:400px"> 
> * Comment lire notre arbre de décision ?
> > 1. Le noeud racine (le premier noeud), contient tous les échantillons du jeu d'entraînement. Ici il y a 50 échantillons : `samples=50`. <br><br>Le premier test qui va être réalisé correspond à `Number of Claims < = 23.5`. **On regarde pour chaque échantillon s'il y a moins de 23.5 réclamations. Si oui, ces échantillons vont dans le noeud enfant de gauche, si non dans le noeud enfant de droite.** <br><br> La valeur moyenne du montant des indemnités (sur tous les échantillons contenus dans `y_train`) est égale à **71.3** : `value = 71.3`. <br><br>  La mesure de l'erreur choisie est ici la "Mean Absolute Error" : `absolute_error = 71.3`. Elle correspond à l'erreur moyenne absolue calculée à partir des valeurs de `y_train` et de la moyenne des valeurs `71.3`. <br> <br> Le calcul se fera ainsi : 
> > > - On retranche `71.3` à chaque valeur contenue dans `y_train`. 
> > >
> > >
> > >
> > > - On calcule la valeur absolue de chacune de ces différences. 
> > >
> > >
> > >
> > > - Enfin, on fait la moyenne de ces valeurs. On obtient alors la `Mean Absolute Error` ou `Erreur Moyenne Absolue`.
> > 
> > A noter que cette mesure de l'erreur est ce qui permet au modèle de sélectionner la variable et le seuil qui va être utilisé dans chaque noeud. On cherche à minimiser cette erreur à chaque étape : la prédiction doit être au plus près des valeurs réelles pour chaque groupe d'échantillons. 
> >
> > 2. Les noeuds enfants sont les noeuds intermédiaires de l'arbre. On les obtient après avoir appliqué le  test présenté dans les noeuds précédents. Ici dans notre exemple, il n'y a qu'un niveau de noeuds intermédiaires après le noeud racine. Dans le noeud enfant gauche, tous les échantillons ont un nombre de réclamations inférieur ou égal à **23.5** et dans celui de droite un nombre supérieur à **23.5**. <br><br>  Nous avons dans notre cas intentionnelement limité la profondeur de l'arbre pour simplifier l'explication mais il pourrait y en avoir largement plus. 
> >
> >
> > 3. Finalement, les feuilles, dites aussi noeuds terminaux, représentent l'issue du modèle. On peut différencier une feuille d'un noeud intermédiaire car elle n'est pas suivie d'un autre noeud. <br><br> On interprète le contenu de la feuille de cette façon, ici on considère la feuille la plus à gauche : elle contient tous les échantillons qui ont un nombre de réclamations inférieur ou égal à **6.5**. Les échantillons qui tombent dans cette feuille sont au nombre de **10**. Le modèle prédit pour ces échantillons un montant de **14.7**. C'est le **montant moyen** des indemnités perçues par ce groupe d'échantillons. 
> >
> 
> * Quelques explications supplémentaires : 
> 
> > 1. On peut se demander comment sont déterminés les nombres au sein des noeuds :`23.5`, `6.5` et `84.5`. Ce sont des **seuils** qui sont obtenus à partir de tests successifs. En effet, de façon itérative, plusieurs seuils vont être testés. Le seuil qui minimise l'erreur (ici la MAE) est celui qui sera retenu. <br><br> **En d'autres termes, on cherche la valeur au sein de la variable qui va nous permettre de séparer nos échantillons le mieux possible. En séparant les groupes d'échantillons, on pourra réaliser une prédiction plus précise de la variable cible pour chacun des groupes. Les groupes d'échantillons finaux sont représentés par les feuilles.** <br><br>
Ainsi dans notre exemple, `23.5` est le seuil qui minimise l'erreur : c'est la première condition que l'on va tester. Puis en considérant cette condition, on procède de la même façon.  Dans le noeud enfant gauche,  on va chercher quel est le seuil inférieur ou égal à 23.5 qui va à nouveau minimiser l'erreur. A droite, on cherchera le seuil, supérieur à 23.5 qui minimisera l'erreur.
> >
> >
> > 2. La couleur dépend de la valeur prédite. Plus la valeur prédite est elevée, plus la couleur est intense.
>
>
> Cette exemple avait vocation à expliquer la lecture d'un arbre de décision. Nous ne rentrons volontairement pas dans le détail de ce cas pratique. Dans le cadre de la création d'un modèle de Machine Learning, il faut bien entendu poursuivre l'étude des performances du modèle en étudiant ses métriques. <h3> 3. Les modèles linéaires </h3> 

> Les modèles linéaires font partie des modèles de Machine Learning les plus simples. C'est ce que nous allons étudier dans cette seconde partie du notebook. 
> 
> #### 3.1 La régression linéaire
>
> Soient X la variable explicative et y la variable cible, obtenus à partir d'un jeu de données. Dans ce cadre, un modèle de régression linéaire **simple** consistera à modéliser le lien entre la variable explicative X et la variable cible y. On prédit donc les valeurs de y à partir des valeurs de X. 
>
> Le modèle peut s'écrire comme une fonction affine : $y=β_0+β_1x+ε$ <br>
> où :   
>> * $y$ est une variable quantitative continue à expliquer (la variable cible).
>>
>>
>> * $x$ est une variable quantitative continue (la variable explicative).
>>
>>
>> * $β_0$ et $β_1$ sont respectivement l'ordonnée à l'origine et le coefficient directeur de la fonction.
>>
>>
>> * $ε$ est un terme d’erreur aléatoire suivant une loi normale d’espérance nulle et d’écart-type σ.
>
> En d'autres termes, cela signifie que les valeurs de y s'obtiennent à partir d'une combinaison linéaire de ces éléments : $β_0$, $β_1$, $x$ et $ε$. 
>
> Néanmoins seules les valeurs de la variable explicative $x$ sont connues. Il faut donc **estimer** les valeurs des paramètres $β_0$ et $β_1$ pour prédire la valeur de y. **C'est ce que l'on fait lorsque l'on entraîne le modèle, on parle également d'estimation.** <br><br> On estimera lors de cette étape les valeurs des coefficients $β_0$ et $β_1$ par $\hat{β}_0$ et $\hat{β}_1$, on aura alors : $\hat{y}=\hat{β}_0+\hat{β}_1x$. où $\hat{y}$ correspond à la prédiction du modèle, $\hat{β}_0$ et $\hat{β}_1$ aux estimations des coefficients du modèle. <br><br> 
> Le terme d'erreur est défini par $ε = y - \hat{y}$. C'est l'écart entre la vraie valeur de y et la valeur prédite. L'objectif lors de l'entraînement du modèle est de réduire au maximum cet écart résiduel. Alors, les paramètres $\hat{β}_0$ et $\hat{β}_1$ choisis seront ceux qui permettent de minimiser l'erreur du modèle.<br><br> 
> Finalement, plus la valeur prédite $\hat{y}$ est proche de la vraie valeur de y, plus le modèle est performant. 
>
> Pour mieux comprendre le fonctionnement de cette étape, vous allez pouvoir chercher à minimiser cette erreur par vous même. Pour évaluer l'erreur du modèle dans sa globalité, on utilise une métrique appelée la Mean Squared Error. A la différence de la Mean Absolute Error, que l'on a évoqué lorsque l'on a analyse l'arbre de décision, plutôt que de calculer la valeur absolue de la différence, on met au carré la différence entre la vraie valeur et la valeur prédite. Ces différences au carré sont ensuite sommées.


> Bien que la visualisation graphique soit très parlante, il n'est pas simple de quantifier les erreurs du modèle. Une solution est de calculer une **métrique de performance**. Il existe plusieurs métriques pour les modèles de régression. Dans le précédent notebook, nous avons présenté le coefficient de détermination, cette métrique s'utilise principalement pour étudier les performances des modèles de régression linéaire et n'est pas adapté à tous les modèles. Nous présentons ici, trois autres métriques utiles : 
>
> - **Mean Absolute Error (MAE)** : C'est la moyenne des différences absolues entre les vraies valeurs et les valeurs prédites du modèle. Elle est facilement interprétable au regard de la variable cible, car elle a la même unité.
>
> <center><video controls src="https://assets-datascientest.s3.eu-west-1.amazonaws.com/notebooks/module_da_121/notebook3/mae_formula.mp4" width = '800' /></center>
> 
>
> - **Mean Squared Error (MSE)** : C'est la moyenne des différences mises au carré entre les vraies valeurs et les valeurs prédites de la variable cible. Elle n'est pas dans la même unité que la variable cible, ce qui rend son interprétation plus complexe. Cette métrique pénalise davantage les grands erreurs que la MAE. Elle est donc intéressante lorsque nous souhaitons absolument éviter ces cas.
>
>
> - **Root Mean Squared Error (RMSE)** : C'est la racine carré de la Mean Squarred Error. Elle pénalise également davantage les grands écarts de prédiction. Elle a la même unité que la variable cible donc elle sera plus interprétable que la MSE.
>
> > <center><video controls src="https://assets-datascientest.s3.eu-west-1.amazonaws.com/notebooks/module_da_121/notebook3/mse_formula.mp4" width = '800' /></center>
>
> <div class="alert alert-warning">
    <i class="fa fa-info-circle"></i> &emsp; <b> Il faut garder en tête que le choix de la métrique peut varier selon les cas de figure. L'analyse de performance des modèles peut différer selon les secteurs et applications. Il n'existe pas une seule métrique optimale. Il est par ailleurs possible de créer une métrique qui répond à nos propres critères, l'objectif étant avec cette mesure de pouvoir décider si le modèle est suffisamment performant pour être retenu dans un cas d'application bien spécifique. </b>
     </div>

>
>> * L'**accuracy**.
>> 
>>
>> * La **précision et le rappel** (*precision* et *recall* en anglais).
>> 
>>
>> * Le **score F1**, ou *F1-score* en anglais.
> 
> Chaque métrique évalue la performance du modèle avec une approche différente.
>
> Afin d'expliquer ces notions, nous allons introduire 4 termes très importants. 
>
>> * **Vrais positifs (VP)** une observation classée **positive** (classe 1) par le modèle et qui est effectivement **positive** (classe 1).
>>
>>
>> * **Faux positifs (FP)** une observation classée **positive** (classe 1) par le modèle mais qui était en réalité **négative** (classe 0).
>>
>>
>> * **Vrais négatifs (VN)** une observation classée **négative** (classe 0) par le modèle et qui est effectivement **négative** (classe 0).
>>
>>
>> * **Faux négatifs (FN)** une observation classée **négative** (classe 0) par le modèle mais qui était en réalité **positive** (classe 1).
>
> <br>
> <img src = "https://assets-datascientest.s3.eu-west-1.amazonaws.com/notebooks/module_da_121/notebook3/confusion_matrix.png" style = "height:300px'">
> <br>
>
> <h5> L'accuracy </h5>
>
> L'**accuracy** est la métrique la plus connue pour évaluer un modèle de classification. <br>
> Elle correspond simplement au taux de prédictions **correctes** effectuées par le modèle. Pour rappel, c'est la métrique qui est utilisée par défaut lorsque l'on utilise la méthode `score`.
>
> On suppose que l'on dispose de $n$ échantillons. <br>
> On note $\mathrm{VP}$ le nombre de Vrais Positifs et $\mathrm{VN}$ le nombre de Vrais Négatifs.<br>
> L'accuracy est alors donnée par :
> $$\mathrm{accuracy} = \frac{\mathrm{VP} + \mathrm{VN}}{n}$$
> 
> La fonction **`accuracy_score`** du sous-module `sklearn.metrics` permet d'obtenir l'accuracy à partir des **prédictions** d'un modèle :
>
> ```python
> accuracy_score(y_true, predictions)
>
> ```
>
>> * **`y_true`** contient les **vraies** valeurs de y.
>>
>>
>> * **`predictions`** contient les valeurs **prédites** par le modèle.
>
> <h5> La précision </h5>
>
> La **précision** est une métrique qui répond à la question : **Parmi toutes les prédictions positives du modèle, combien sont de vrais positifs ?**
>
> Si on note $\mathrm{FP}$ le nombre de Faux Positifs du modèle, alors la précision est donnée par :
> $$\mathrm{precision} = \frac{\mathrm{VP}}{\mathrm{VP} + \mathrm{FP}}$$
>
> Un score de précision élevé nous informe que le modèle ne classe pas aveuglément toutes les observations comme positives.
>
> La fonction **`precision_score`** du sous-module `sklearn.metrics` permet d'obtenir le rappel à partir des **prédictions** d'un modèle :
>
> ```python
> precision_score(y_true, predictions)
>
> ```
>
>> * **`y_true`** contient les **vraies** valeurs de y.
>>
>>
>> * **`predictions`** contient les valeurs **prédites** par le modèle.
> 
> <h5> Le rappel </h5>
>
> Le **rappel** est une métrique qui quantifie la proportion d'observations réellement positives qui ont été correctement classifiées positives par le modèle.
>
> Si on note $\mathrm{FN}$ le nombre de Faux Négatifs, alors le rappel est donné par :
> $$\mathrm{rappel} = \frac{\mathrm{VP}}{\mathrm{VP} + \mathrm{FN}}$$
>
> Un score de rappel élevé nous informe que le modèle est capable de bien détecter les observations réellement positives.
>
> La fonction **`recall_score`** du sous-module `sklearn.metrics` permet d'obtenir le rappel à partir des **prédictions** d'un modèle :
>
> ```python
> recall_score(y_true, predictions)
>
> ```
>
>> * **`y_true`** contient les **vraies** valeurs de y.
>>
>>
>> * **`predictions`** contient les valeurs **prédites** par le modèle.
>
> <h5> La matrice de confusion </h5>
>
> La **matrice de confusion** compte pour un jeu de données les valeurs de VP, VN, FP et FN, ce qui nous permet de calculer les trois métriques précédentes :
>
> $$
\mbox{Confusion Matrix} = \begin{bmatrix}
                                    \mathrm{VN} & \mathrm{FP} \\
                                    \mathrm{FN} & \mathrm{VP}
                                \end{bmatrix}
 $$
>
> La fonction **`confusion_matrix`** du sous-module `sklearn.metrics` permet de générer la matrice de confusion à partir des **prédictions** d'un modèle :
>
> ```python
> confusion_matrix(y_true, predictions)
>
> ```
>
>> * **`y_true`** contient les **vraies** valeurs de y.
>>
>>
>> * **`predictions`** contient les valeurs **prédites** par le modèle.
>

* **(e)** Créez une fonction qui retourne l'accuracy, la précision, le rappel et la matrice de confusion à partir des prédictions de notre modèle sur le jeu de test.


> L'interprétation des coefficients de la régression logistique est néanmoins un petit peu différente que celle de la régression linéaire. Il faut interpréter les coefficients en calculant leurs **odd-ratios**. Ils correspondent à un **rapport des chances**. <br><br> Les odd-ratios sont obtenus en calculant l'exponentielle des coefficients. Par exemple, après avoir estimé β1, l'odd-ratio sera égal à $e^{β1}$.  <br><br> *Mais comment s'interprète cette valeurs ?* <br><br>
Dans notre cas, nous avons estimé la valeur de β1 à **3.61**. Son odd ratio sera égal à **37**. Cela peut s'interpréter ainsi : lorsque la variable explicative augmente d'**une unité**, alors les **chances d'appartenir à la classe 1 augmentent de 3600%.** 
> Plus généralement, si l'odd ratio est supérieur à 1, cela signifie qu'une augmentation d'une unité de la variable explicative augmente les chances d'appartenir à la classe positive. S'il est égal à 1, la variable n'a pas d'impact sur la variable cible et finalement s'il est inférieur à 1, une augmentation d'une unité de de la variable diminue les chances d'appartenir à la classe positive.<br><br>
> Par exemple si un **odd ratio égal à 1.2**, cela signifie que les chances d'appartenir à la classe 1 augmentent de 20%. Lorsqu'un **odd ratio est égal à 2**, les **chances d'appartenir à la classe 1 augmentent de 100%**, un **odd ratio de 3 de 200% etc**. Au contraire, pour un **odd ratio égal à 0.2**, **les chances d'appartenir à la classe 1 diminuent de 20%** et ainsi de suite.
